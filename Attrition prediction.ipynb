### feature engg

```python
from sklearn.preprocessing import LabelEncoder # import Labelencoder 
```

```python
 # covert categorical data into numarical
for i in df.columns:
    if df[i].dtype == np.number:
        continue
    else:
        df[i] = LabelEncoder().fit_transform(df[i])
```

```python
df.head()
```

### Model building

```python
#splite data into feature and target matrx
feature_mtx = df.drop('Attrition',axis = 1)
target_mtx = df['Attrition']
```

```python
feature_mtx.head()
```

```python
target_mtx
```

```python
from sklearn.model_selection import train_test_split
```

```python
#spilte data for training and testing
x_train,x_test,y_train,y_test = train_test_split(feature_mtx,target_mtx,test_size = 0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape #cheak shape 
```

```python
# import RandomForestClassifier to train model
from sklearn.ensemble import RandomForestClassifier
```

```python
# fit model on randomfroest
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)*100
```

```python
#predict from trained model 
y_pred = rf.predict(x_test)
```

```python
# to check model 
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
print('accuracy',accuracy_score(y_pred,y_test)*100)
print('precision_score',precision_score(y_pred,y_test))
print('precision_score',recall_score(y_pred,y_test))
print('precision_score',f1_score(y_pred,y_test))
print('confusion_matrix')
print(confusion_matrix(y_pred,y_test))
```

```python
sns.heatmap(confusion_matrix(y_pred,y_test))
```
